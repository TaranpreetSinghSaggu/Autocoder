```python
class PrimKruskalMST:
    def __init__(self, vertices):
        self.vertices = vertices
        self.edges = []

    def add_edge(self, u, v, weight):
        self.edges.append((u, v, weight))

    def find(self, parent, i):
        if parent[i] == i:
            return i
        return self.find(parent, parent[i])

    def union(self, parent, rank, x, y):
        xroot = self.find(parent, x)
        yroot = self.find(parent, y)
        if rank[xroot] < rank[yroot]:
            parent[xroot] = yroot
        elif rank[xroot] > rank[yroot]:
            parent[yroot] = xroot
        else:
            parent[yroot] = xroot
            rank[xroot] += 1

    def kruskal_mst(self):
        mst = []
        edges = sorted(self.edges, key=lambda x: x[2])
        parent = {}
        rank = {}

        for vertex in self.vertices:
            parent[vertex] = vertex
            rank[vertex] = 0

        for u, v, weight in edges:
            x = self.find(parent, u)
            y = self.find(parent, v)
            if x != y:
                mst.append((u, v, weight))
                self.union(parent, rank, x, y)

        return mst
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# generate random data
np.random.seed(42)
data = pd.DataFrame({'Age': np.random.randint(18, 61, 100), 'Salary': np.random.randint(40000, 120000, 100)})

# group by age decade
grouped = data.groupby(lambda x: x.age // 10)

# calculate average salary per age decade
average_salary = grouped['Salary'].mean()

# plot bar chart of average salary
average_salary.plot(kind='bar')
plt.title('Average Salary by Age Decade')
plt.xlabel('Age Decade')
plt.ylabel('Average Salary')
plt.show()
```
